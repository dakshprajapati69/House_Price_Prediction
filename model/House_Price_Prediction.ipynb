```python
def is_float(x):
    try:
        float(x)
    except:
        return False
    return True
    
```

```python
df[~df['total_sqft'].apply(is_float)].head()
```

```python
def convert_num_total_sqft(x):
    try:
        tokens = str(x).split('-')   # str(x) handles NaN/float values safely
        if len(tokens) == 2:
            return (float(tokens[0]) + float(tokens[1])) / 2
        return float(x)
    except:
        return None  # handles anything non-convertible

df['total_sqft'] = df['total_sqft'].apply(convert_num_total_sqft)
df.dropna(subset=['total_sqft'], inplace=True)
df['total_sqft'].dtype  # should be float64 now
```

```python
df['total_price_per_sqft'] = (df['price']*100000) / df['total_sqft']
df.head()
```

```python
df.location = df.location.apply(lambda x :'other' if x in state_less_than_10 else x)
len(df.location.unique())
```

```python
def outlier_ppp_removel(df):
    df_out = pd.DataFrame()
    for key,subdf in df.groupby('location'):
        m = np.mean(subdf.total_price_per_sqft)
        sd = np.std(subdf.total_price_per_sqft)
        reduce_df = subdf[(subdf.total_price_per_sqft>(m-sd)) & (subdf.total_price_per_sqft<=(m+sd))]
        df_out = pd.concat([df_out,reduce_df],ignore_index=False)

    return df_out

df = outlier_ppp_removel(df)
df.shape
    
```

```python
def remove_bhk_outliers(df):
    exclude_indices = np.array([])
    
    for location, location_df in df.groupby('location'):
        bhk_stats = {}
        
        for bhk, bhk_df in location_df.groupby('BHK'):
            bhk_stats[bhk] = {
                'mean': np.mean(bhk_df.total_price_per_sqft),
                'std': np.std(bhk_df.total_price_per_sqft),
                'count': bhk_df.shape[0]
            }
        
        for bhk, bhk_df in location_df.groupby('BHK'):
            stats = bhk_stats.get(bhk - 1)
            if stats and stats['count'] > 5:
                exclude_indices = np.append(
                    exclude_indices,
                    bhk_df[bhk_df.total_price_per_sqft < (stats['mean'])].index.values
                )
    
    return df.drop(exclude_indices, axis='index')

df = remove_bhk_outliers(df)
df.shape
```

```python
from sklearn.linear_model import Lasso
from sklearn.model_selection import GridSearchCV
from sklearn.tree import DecisionTreeRegressor
```

```python
def find_best_model(X,y):
    algos = {
        'Linear Regression' : {
            'model' : LinearRegression(),
            'params' : {
                
            }
        },
        'Lasso' : {
            'model' : Lasso(),
            'params' : {
                'alpha' : [1,2],
                'selection' : ['random','cyclic']
            }
        },
        'Decision Tree' : {
            'model' : DecisionTreeRegressor(),
            'params' : {
                'splitter' : ['best','random'],
                'criterion' : ['squared_error','friedman_mse']
            }
        }
    }

    scores = []
    cv = ShuffleSplit(n_splits=5,test_size=0.2,random_state=False)
    for algo_name,config in algos.items():
        grid = GridSearchCV(config['model'],config['params'],cv=cv,return_train_score=False)
        grid.fit(X,y)
        scores.append(
           { 'model': algo_name,
            'Best Score' : grid.best_score_,
            'Best Params' : grid.best_params_  
    })

    return pd.DataFrame(scores,columns=['model','Best Score','Best Params'])

find_best_model(X,y)
```

```python
def Predict_price(location,sqft,bath,bhk):
    loc_idx = np.where(X.columns==location)[0][0]

    x = np.zeros(len(X.columns))
    x[0] = sqft
    x[1] = bath
    x[2] = bhk

    if loc_idx >= 0:
        x[loc_idx] = 1

        

    return model.predict([x])
```
